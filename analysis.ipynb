```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sales = pd.read_csv("sales_data.csv")
df_sales.head()
```

```python
df_sales.info()
df_sales.describe()
```

```python
# ===============================
# 1. Bar Chart: Sales by Product Category
# ===============================
sales_data = {
    "Category": ["Electronics", "Clothing", "Grocery", "Furniture"],
    "Sales": [120000, 90000, 150000, 70000]
}

df_sales = pd.DataFrame(sales_data)

plt.figure()
plt.bar(df_sales["Category"], df_sales["Sales"])
plt.title("Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Sales Amount")
plt.show()
```

```python
weather_df = pd.read_csv("weather_data.csv")
weather_df.head()
```

```python
weather_df.dtypes
```

```python
# ===============================
# 2. Line Chart: Monthly Temperature changes
# ===============================
weather_df["Date_Time"] = pd.to_datetime(weather_df["Date_Time"], errors="coerce")
weather_df = weather_df.dropna(subset=["Date_Time"])

monthly_avg = (
    weather_df
    .assign(Month=weather_df["Date_Time"].dt.to_period("M"))
    .groupby("Month")["Temperature_C"]
    .mean()
)

plt.figure(figsize=(10, 5))
monthly_avg.plot()
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
# ===============================
# 3. Pie Chart: Expense Distribution
# ===============================
expenses = {
    "Rent": 40,
    "Food": 25,
    "Transport": 15,
    "Utilities": 10,
    "Entertainment": 10
}

plt.figure()
plt.pie(
    expenses.values(),
    labels=expenses.keys(),
    autopct="%1.1f%%"
)
plt.title("Monthly Expense Distribution")
plt.show()
```

```python
# ===============================
# 4. Data Analysis Summary
# ===============================
print("Total Sales:", df_sales["Sales"].sum())
print("Highest Selling Category:", df_sales.loc[df_sales["Sales"].idxmax(), "Category"])
print("Average Temperature:", weather_df["Temperature_C"].mean())
```
